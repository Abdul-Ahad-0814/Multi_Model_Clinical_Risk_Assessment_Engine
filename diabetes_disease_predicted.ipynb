```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df_diabetes= pd.read_csv("diabetes.csv")
df_diabetes.head()
```

```python
df_diabetes= df_diabetes.rename(columns={
    "Age": "age",
    "BloodPressure": "systolic_bp",
    "Glucose": "glucose",
    "BMI": "bmi",
    "Outcome": "target"
})
df_diabetes.head()
```

```python
df_diabetes['gender']= 0
df_diabetes['diastolic_bp']= 80
df_diabetes['cholesterol']= 200
df_diabetes['smoking']= 0
df_diabetes['alcohol']= 0
df_diabetes['family_history']=0
df_diabetes.head()
```

```python
schema_cols = [
    "age",
    "gender",
    "systolic_bp",
    "diastolic_bp",
    "glucose",
    "cholesterol",
    "bmi",
    "smoking",
    "alcohol",
    "family_history",
    "target"
]

df_diabetes= df_diabetes[schema_cols]
df_diabetes.head()
```

```python
x= df_diabetes.drop("target", axis=1)
y= df_diabetes["target"]
```

```python
from sklearn.pipeline import make_pipeline

diabetes_stage2= make_pipeline(
    StandardScaler(),
    LogisticRegression()
)
```

```python
diabetes_stage1= make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, class_weight={0:1,1:2})
)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2, random_state=106)
# x_train_scaled= StandardScaler().fit_transform(x_train)
# x_test_scaled= StandardScaler().fit_transform(x_test)
diabetes_stage1.fit(x_train,y_train)
y_pred= diabetes_stage1.predict(x_test)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
